```python
def analytic(p):
    g = 0.04
    t = (p/p[-1])*256
    sb = 5.67e-8
    fint = sb*70**4
    s = 1368/4
    
    t4 = fint*(t+1) + s*(1 + 1/g + (g - 1/g)*np.exp(-g*t))
    return (t4/2/sb)**(1/4)
```
